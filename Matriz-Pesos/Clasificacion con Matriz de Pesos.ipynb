```python
df = sns.load_dataset('iris')
df.head()
```

```python
#Bias
#Sepal Length
#Sepal Width
#Petal Length
#Petal Width
```

```python

entradas = df[['sepal_length','sepal_width', 'petal_length','petal_width']].copy()
entradas.insert(0, "b", 1, allow_duplicates=False)
entradas
```

```python
capa1= np.dot(entradas,Pesos1)
capa1 = pd.DataFrame(capa1)
for i in range(len(capa1)):
    for j in range(capa1.shape[1]):
        capa1.iloc[i][j]  = (1/(1+(np.exp(-capa1.iloc[i][j]))))
capa1.insert(0, "b", 1, allow_duplicates=False)
capa1
```

```python
capa2= np.dot(capa1,Pesos2)
capa2 = pd.DataFrame(capa2)
for i in range(len(capa2)):
    for j in range(capa2.shape[1]):
        capa2.iloc[i][j]  = (1/(1+(np.exp(-capa2.iloc[i][j]))))
capa2.insert(0, "b", 1, allow_duplicates=False)
capa2
```

```python
capa3= np.dot(capa2,Pesos3)
capa3 = pd.DataFrame(capa3)
capa3
```

```python
resultados = []
MaxId = pd.DataFrame()
MaxId["Max"] = capa3.idxmax(axis=1)
Species = pd.DataFrame()
for i in range(len(MaxId)):
    valor = MaxId['Max'][i]
    if (valor == 0):
        resultados.append("setosa")
    elif (valor ==1):
        resultados.append("versicolor")
    else:
        resultados.append("virginica")
```

```python
df["Clasificacion"]= resultados
df.to_csv('Resultados.csv', sep=";" ,index=False)
#Comparacion de clasificacion
df
```
